<font color='Blue'>*Во-первых.*</font>
Пусть последовательность $(p_n)_{n \geqslant 1}$ чисел $p_n\in(0, 1)$ такова, что $np_n\to\lambda>0$.

<font color='Blue'>*Во-вторых.*</font>
На измеримом пространстве $(\mathbb{N}, \mathscr{P}(\mathbb{N}))$ определим последовательность мер $\mathsf{P}_n$ посредством задания их плотностей относительно считающей меры:

$$
		\frac{\mathrm{d}\mathsf{P}_n}{\mathrm{d}\#}(x) =
		\begin{cases}
			C_n^xp_n^x(1-p_n)^{1-x},
			&\text{если } 0 \leqslant x \leqslant n, \\
			0, &\text{иначе}.
		\end{cases}
$$

<font color='Blue'>**Теорема Пуассона**</font> утверждает, что $\mathsf{P}_n\to\mathrm{Pois}(\lambda)$. Напомню, что

$$
    \frac{\mathrm{d}\mathrm{Pois}(\lambda)}{\mathrm{d}\#}(x) =
    \frac{\lambda^x}{x!}e^{-\lambda}.
$$

```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
```

```python
probabilities = []
for n in range(1, 16):
    p_n = 11 / (12 + n)
    probabilities.append(p_n)
probabilities = np.array(probabilities)
probabilities
```

```python
arguments = np.arange(0, 31)
```

```python
for n in range(1, 15):
    p_n = probabilities[n-1]
    expectation = n * p_n
    print(f'Число испытаний: \t\t {n}')
    print(f'Вероятность успеха: \t\t {p_n:.2f}')
    print(f'Математическое ожидание: \t {expectation:.2f}')
    pmfValues = binom.pmf(arguments,n,p_n)
    plt.figure(figsize=(20,6))
    plt.plot(arguments,pmfValues,'bo',ms=15)
    plt.vlines(arguments, 0, pmfValues, colors='b', lw=15, alpha=0.4)
    plt.title(f'Плотность биномиального распределения (по считающей мере) с \
вектором параметров ({n}, {p_n:.2f})', fontsize=20)
    plt.xlabel('Натуральное число $x$', fontsize=14)
    plt.ylabel('PMF($x$)',fontsize=14)
    plt.tick_params(axis='both',which='major',labelsize=14)
    plt.show()
```
