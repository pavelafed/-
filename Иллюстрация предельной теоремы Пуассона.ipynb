# Первая точка зрения на теорему Пуассона.

<font color='Blue'>**1.**</font>
Пусть последовательность $(p_n)_{n \geqslant 1}$ чисел $p_n\in(0, 1)$ такова, что $np_n\to\lambda>0$.

<font color='Blue'>**2.**</font>
На измеримом пространстве $(\mathbb{N}, \mathscr{P}(\mathbb{N}))$ определим последовательность мер $\mathsf{P}_n$ посредством задания их плотностей относительно считающей меры:

$$
		\frac{\mathrm{d}\mathsf{P}_n}{\mathrm{d}\#}(x) =
		\begin{cases}
			C_n^xp_n^x(1-p_n)^{1-x},
			&\text{если } 0 \leqslant x \leqslant n, \\
			0, &\text{иначе}.
		\end{cases}
$$

<font color='Blue'>**Теорема Пуассона**</font> говорит о ***поточечной*** сходимости мер <font color='Red'>$\mathrm{Bin}(n, p_n)\to\mathrm{Pois}(\lambda)$</font>. Напомню, что

$$
    \frac{\mathrm{d}\mathrm{Pois}(\lambda)}{\mathrm{d}\#}(x) =
    \frac{\lambda^x}{x!}e^{-\lambda}.
$$

# Вторая точка зрения на теорему Пуассона.

<font color='Blue'>**1.**</font>
Пусть последовательность $(p_n)_{n \geqslant 1}$ чисел $p_n\in(0, 1)$ такова, что $np_n\to\lambda>0$.

<font color='Blue'>**2.**</font>
На измеримом пространстве $(\mathbb{R}, \mathscr{B}(\mathbb{R}))$ определим последовательность ***дискретных*** мер $\mathrm{Bin}(n, p_n)$ посредством задания последовательности $(m_n)_{n\geqslant1}$ нагрузок:

$$
    m_n\colon\mathbb{R}\to\mathbb{R}
    \quad::\quad
    x\mapsto
    \begin{cases}
        C_n^x p_n^x (1-p_n)^{1-x}, &\text{если } x = 0, 1, ..., n; \\
        0, &\text{иначе};
    \end{cases}
$$

$$
		\mathrm{Bin}(n, p_n)\colon\mathscr{B}(\mathbb{R})\to[0, 1]
    \quad::\quad
    B \mapsto \sum_{x \in B} m_n(x).
$$

Здесь мы использовали т.н. суммируемые семейства ([ссылка](https://youtu.be/9Owx8Ne-d6o?list=PL-_cKNuVAYAU1iqXkA388zr0c45dKvN5Q&t=1128) на лекции С. В. Кислякова по этой теме, матмех СПбГУ).

<font color='Blue'>**Теорема Пуассона**</font> говорит о ***слабой*** сходимости мер <font color='Red'>$\mathrm{Bin}(n, p_n)\rightharpoonup\mathrm{Pois}(\lambda)$</font>, где

$$
    m\colon\mathbb{R}\to\mathbb{R}
    \quad::\quad
    x \mapsto
    \begin{cases}
        \dfrac{\lambda^x}{x!}e^{-\lambda},
        &\text{если } x\in\mathbb{N} \cup \{0\}; \\
        0, &\text{иначе};
    \end{cases}
$$

$$
    \mathrm{Pois}(\lambda)\colon\mathscr{B}(\mathbb{R})\to[0, 1]
    \quad::\quad
    B \mapsto \sum_{x \in B} m_n(x)..
$$

# Иллюстрация предельной теоремы Пуассона.

```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
```

```python
probabilities = []
for n in range(1, 16):
    p_n = 11 / (11 + n)
    probabilities.append(p_n)
probabilities = np.array(probabilities)
probabilities
```

```python
arguments = np.arange(0, 31)
```

```python
for n in range(1, 15):
    p_n = probabilities[n-1]
    expectation = n * p_n
    print(f'Число испытаний: \t\t {n}')
    print(f'Вероятность успеха: \t\t {p_n:.2f}')
    print(f'Математическое ожидание: \t {expectation:.2f}')
    pmfValues = binom.pmf(arguments,n,p_n)
    plt.figure(figsize=(20,6))
    plt.plot(arguments,pmfValues,'bo',ms=15)
    plt.vlines(arguments, 0, pmfValues, colors='b', lw=15, alpha=0.4)
    plt.title(f'Плотность биномиального распределения (по считающей мере) с \
вектором параметров ({n}, {p_n:.2f})', fontsize=20)
    plt.xlabel('Натуральное число $x$', fontsize=14)
    plt.ylabel('PMF($x$)',fontsize=14)
    plt.tick_params(axis='both',which='major',labelsize=14)
    plt.show()
```

```python
from scipy.stats import poisson
poisson.pmf(arguments,11)
```

```python
pmfValues = poisson.pmf(arguments,11)
plt.figure(figsize=(20,6))
plt.plot(arguments,pmfValues,'bo',ms=15)
plt.vlines(arguments, 0, pmfValues, colors='b', lw=15, alpha=0.4)
plt.title(f'Плотность распределения Пуассона (по считающей мере) с \
параметром 11', fontsize=20)
plt.xlabel('Натуральное число $x$', fontsize=14)
plt.ylabel('PMF($x$)',fontsize=14)
plt.tick_params(axis='both',which='major',labelsize=14)
plt.show()
```
