```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
```

```python
probabilities = []
for n in range(1, 21):
    p_n = 11 / (12 + n)
    probabilities.append(p_n)
probabilities
```

```python
arguments = np.arange(0,21)

for n in range(1, 21):
    p_n = probabilities[n-1]
    expectation = n * p_n
    pmfValues = binom.pmf(arguments,n,p_n)
    plt.figure(figsize=(20,6))
    plt.plot(arguments,pmfValues,'bo',ms=15)
    plt.vlines(arguments, 0, pmfValues, colors='b', lw=15, alpha=0.4)
    plt.text(15, 0.1, f'Математическое ожидание: {expectation:.2f}', fontsize=14)
    plt.title(f'Плотность биномиального распределения (по считающей мере) с \
вектором параметров ({n}, {p_n:.2f})', fontsize=20)
    plt.xlabel('Натуральное число $x$', fontsize=14)
    plt.ylabel('PMF($x$)',fontsize=14)
    plt.tick_params(axis='both',which='major',labelsize=14)
    plt.show()
```
